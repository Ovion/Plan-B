```python
import pandas as pd
```

```python
bici = pd.read_csv('bici.csv')
```

```python
bici.fecha = pd.to_datetime(bici.fecha, yearfirst=True)
bici.dia = pd.to_datetime(bici.dia, yearfirst=True)
```

```python
bici['year']=bici.dia.dt.year
```

```python
bici.dropna(inplace=True)
```

```python
bici= bici[['fecha','dia', 'year', 'horario', 'festividad', 'tipo_accidente', 'lesividad', 'meteo', 'distrito', 'direccion', 'lon', 'lat']]
```

```python
bici.festividad.value_counts()
```

```python
for i in range(0, bici.shape[0]-1):
    if (bici.iat[i, bici.columns.get_loc('festividad')]=='Laborable' and bici.iat[i+1, bici.columns.get_loc('festividad')]=='Festivo'):
        bici.iat[i, bici.columns.get_loc('festividad')] = 'Vispera'
        
```
